Add column

```python
canada[years].sum(axis=1) # Total immigration for each country 
```

```python
canada['Total'] = canada[years].sum(axis=1) # a new column 'Total' will be added to end of dataframe
```

```python
canada.head(2)
```

Remove column

```python
cols_to_drop=['Type','Coverage','AREA','REG','DEV']
canada.drop(columns=cols_to_drop,inplace=True)
canada
```

```python
canada.drop_duplicates(subset=['AreaName'])
```

Renaming columns

```python
canada.sample()
```

```python
canada.rename(mapper={
    'AreaName' : 'Continent' ,
    'RegName' : 'Region' ,
    'DevName' : 'Status'
}, axis=1, inplace=True)
```

```python
canada.head(2)
```

Reordering the column sequence

```python
updated_list= ['Status','Continent','Region','Total']+years
print(updated_list)
```

```python
canada=canada[updated_list] # reordering columns
```

```python
canada.head(2)
```

Grouping

```python
canada.groupby('Status').get_group('Developing regions')
```

```python
canada.groupby('Continent').get_group('Europe')
```

```python
canada.groupby('Status')[[2000,2010]].sum().plot(kind='pie',subplots=True,autopct='%1.1f%%',figsize=(5,10),layout=(2,1))
                                                
```

```python
canada.groupby('Continent')['Total'].mean()
```

```python
canada.groupby('Continent')['Total'].agg(['mean','median'])
```

```python
canada.groupby(['Continent','Region'])['Total'].sum().reset_index()
```

Pivot

```python
canada.pivot_table(columns='Status',index='Continent',values='Total',aggfunc=['min','max'])
```

```python
penguins=sns.load_dataset('penguins')
penguins.columns
```

```python
penguins_pivot=penguins.pivot_table(index='species',columns='sex',values='flipper_length_mm')
penguins_pivot
```

```python
penguins_pivot.plot(kind='bar',stacked=True,figsize=(10,5))
```

```python
penguins_pivot.replace(np.nan,0,inplace=True)
```
